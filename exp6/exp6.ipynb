```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

data_near = np.genfromtxt('exp6_data_near.csv',delimiter=',') # importing data_al files

d = 9e-3
bc_freq = 135/300

def fitfunc(x,a,b,c):
    return a*(1-(x/(np.sqrt(x**2+b**2))))+c

distance = d*data_near[:,1]
freqs = data_near[:,2]/data_near[:,0]

n_params, n_cov = curve_fit(fitfunc,distance,freqs)

plt.style.use(plt.style.available[25])
#plt.rcParams['text.usetex'] = True
plt.plot(distance,freqs,label=r'data')
plt.plot(distance,fitfunc(distance,*n_params),label=r"fitted function")

plt.title(r'near field data')
plt.xlabel(r'thickness*density} (g/cm^2)')
plt.ylabel(r'counts in 60s')
#plt.text(0.04,500,r'R: %.4f cm^2/gr$' % n_err,
#         bbox={'facecolor':'white','ec':'none'})
plt.ylim(bottom=0)
plt.xlim(right=1.1*distance.max())
plt.minorticks_on()
plt.grid(True,which='both')
lg = plt.legend(frameon=True)
lg.get_frame().set_linewidth(0.0)
plt.savefig('exp6_plot_near.jpg',dpi=500)
plt.show()
```

```python
data_far = np.genfromtxt('exp6_data_far.csv',delimiter=',') # importing data_al files

d_f = 1e-2
bc_freq_f = 122/300


distance_f = d_f*data_far[:,1]
freqs_f = data_far[:,2]/data_far[:,0]


f_params, f_cov = curve_fit(fitfunc,distance_f,freqs_f)

plt.style.use(plt.style.available[25])
plt.plot(distance_f,freqs_f,label=r'data')
plt.plot(distance_f,fitfunc(distance_f,*f_params),label=r"fitted function")
plt.title(r'far field data')
plt.xlabel(r'thickness*density (g/cm^2)')
plt.ylabel(r'counts in 60s')
#plt.text(1.5,3,r'R: %.4f cm^2/gr$' % f_err,
#         bbox={'facecolor':'white','ec':'none'})
plt.ylim(bottom=0)
plt.xlim(right=1.1*distance_f.max())
plt.minorticks_on()
plt.grid(True,which='both')
lg = plt.legend(frameon=True)
lg.get_frame().set_linewidth(0.0)
plt.savefig('exp6_plot_far.jpg',dpi=500)
plt.show()
```
